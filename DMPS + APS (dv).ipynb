```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Excel file
file_path = '/Users/touqeergill/Desktop/2. DMPS + APS Hourly Avg (2010-2022) copy.xlsx'
df = pd.read_excel(file_path)

# Convert the Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Define a function to assign seasons
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

# Assign seasons
df['Season'] = df['Date'].dt.month.apply(get_season)

# Extract particle diameter columns and convert to numeric values
diameter_cols = df.columns[1:-1]  # Exclude Date and Season
diameters = np.array([float(col) for col in diameter_cols])

# Calculate mean size distribution per season
seasonal_means = df.groupby('Season')[diameter_cols].mean()

# Plotting
plt.figure(figsize=(10, 6))
colors = {'Winter': 'navy', 'Spring': 'green', 'Summer': 'purple', 'Autumn': 'darkorange'}

for season in ['Winter', 'Spring', 'Summer', 'Autumn']:
    plt.plot(diameters, seasonal_means.loc[season], label=season, color=colors[season], linewidth=2)

plt.xscale('log')
plt.xlabel('$D_p$ (m)', fontsize=12)
plt.ylabel(r'dV/dlog$D_p$ ($\mu$m$^3$ cm$^{-3}$)', fontsize=12)
plt.title('Seasonal Mean Size Distribution', fontsize=14)

# Vertical lines for particle mode regions
plt.axvline(1e-8, color='k', linestyle='--', linewidth=0.8)
plt.axvline(1e-7, color='k', linestyle='--', linewidth=0.8)
plt.axvline(1e-6, color='k', linestyle='--', linewidth=0.8)

# Text annotations
plt.text(3e-9, max(seasonal_means.max())*0.9, 'Nucleation\nmode', fontsize=10)
plt.text(3e-8, max(seasonal_means.max())*0.9, 'Aitken\nmode', fontsize=10)
plt.text(3e-7, max(seasonal_means.max())*0.9, 'Accumulation\nmode', fontsize=10)
plt.text(3e-6, max(seasonal_means.max())*0.9, 'Coarse\nparticles', fontsize=10)

plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
